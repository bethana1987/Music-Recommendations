# Pre processing for df2 

```python
df2.head()
```

```python
# we are going to create the variable "sentiment_score" which will have two values 1 and 0 to predict whether a piece of music 
#will be more or less appreciated based on "vader_avg"
df2['sentiment_score'] = np.where(df2['vader_avg']>= 0.5, 1, 0)
df2.head()
```

-At this point "hashtag" is no longer important as it is represented quantitatively by "vader_avg"

- "created_at", "artist_id" dont provide relevant  information 

- "tweet lang" is not relevant anymore as it is represented by "vader_avg"

- "lang" is not longer relevant as it is represented by English language 

```python
df2_new = df2.drop(['hashtag','created_at','artist_id','tweet_lang','lang'], axis=1)

df2_new.head()
```

```python
df2_new.time_zone.unique()
```

```python
#Make all USA Timezone values consistent
df2_new['time_zone'].replace('Eastern Time (US & Canada)', 'Eastern Time',inplace=True)
df2_new['time_zone'].replace('Central Time (US & Canada)', 'Central Time',inplace=True)
df2_new['time_zone'].replace('Pacific Time (US & Canada)', 'Pacific Time',inplace=True)
df2_new['time_zone'].replace('Mountain Time (US & Canada)', 'Mountain Time',inplace=True)
df2_new['time_zone'].replace('Alaska', 'Alaska Time',inplace=True)
df2_new['time_zone'].replace('Hawaii', 'Hawaii Time',inplace=True)
df2_new['time_zone'].replace('Arizona', 'Mountain Time',inplace=True)
df2_new['time_zone'].replace('America/Chicago', 'Central Time',inplace=True)
df2_new['time_zone'].replace('America/New_York', 'Eastern Time',inplace=True)
df2_new['time_zone'].replace('America/Los_Angeles', 'Pacific Time',inplace=True)
df2_new['time_zone'].replace('America/Denver', 'Mountain Time',inplace=True)
df2_new['time_zone'].replace('America/Detroit', 'Eastern Time',inplace=True)
```

```python
#Limit dataset to only USA time zones
df2_new_tz= df2_new.loc[~((df2_new['time_zone'] != 'Eastern Time') & (df2_new['time_zone'] != 'Central Time') & (df2_new['time_zone'] != 'Pacific Time') & 
               (df2_new['time_zone'] != 'Mountain Time') & (df2_new['time_zone'] != 'Alaska Time') & (df2_new['time_zone'] != 'Hawaii Time')),:]
df2_new_tz.info()
```

```python
value_counts = df2_new_tz['time_zone'].value_counts()
print(value_counts)
```

```python
df2_new_tz_dic=df2_new_tz.join(pd.get_dummies(df2_new_tz.time_zone, prefix="tz"))
df2_new_tz_dic.head()
```
